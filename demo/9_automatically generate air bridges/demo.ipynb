```python
import sys
sys.path.append("../..")
from api.design import Design
from addict import Dict
```

```python
def add_control_lines(design, pos, name="charge_line1", line_type="ControlLineCircle2412", chip_type="chip0"):
    """
    添加控制线到设计中
    """
    design.gds.control_lines.add(options=Dict(
        name=name,
        type=line_type,
        chip=chip_type,
        pos=pos,
        corner_radius=80,
        end_radius=10
    ))
design = Design()

pos = [(0, 0), (200, 0), (200, 550), (800, 550), (500, 0), (900, 0), (1500, 550)]

add_control_lines(design, pos)

spacing = 120
design.gds.auto_generate_air_bridge4(line_type="control_lines",
                                     line_name="charge_line1",
                                     spacing=spacing,
                                     chip_name="chip0",
                                     width=8,
                                     air_bridge_type="AirbridgeNb")
# design.gds.auto_generate_air_bridge2(design.gds.control_lines.charge_line1.pos,design.gds.control_lines.charge_line1.corner_radius,spacing,width=8)
design.gds.show_svg()
```

```python
airbridge_count = sum(1 for item in design.gds.options['air_bridges'].values() if item['type'] == 'AirbridgeNb')
print("空气桥个数：",airbridge_count)
```
